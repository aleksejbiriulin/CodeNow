Удаляем столбцы которые не имеют большой значимости

```python
from sklearn.model_selection import train_test_split
y = df_dummy["Mean Score"]
X = df_dummy.drop("Mean Score", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

Разделили выборку на обучающую и тестовую. При этом мы не использовали нормализацию так как распределение баллов за тесты близко к нормальному.

За базу возьмём просто среднее значение среднего балла по всей выборке.

```python
from sklearn.metrics import mean_squared_error
import numpy as np
y_pred = np.full(len(y_test), y_test.mean())
mean_squared_error(y_test, y_pred)
```

### Обучение модели линейной регрессии и оценка её производительности

В этом разделе мы будем обучать модель линейной регрессии с использованием библиотеки scikit-learn. Линейная регрессия является одним из самых простых и популярных методов для решения задач регрессии. Она основывается на предположении, что существует линейная зависимость между входными переменными (фичами) и целевой переменной.

1. **Создание модели**:
   - Мы начинаем с создания экземпляра модели LinearRegression из модуля linear_model.

2. **Обучение модели**:
   - Затем мы обучаем модель на обучающей выборке, используя метод fit, который принимает матрицу признаков (X_train) и вектор целевых значений (y_train).

3. **Прогнозирование**:
   - После обучения модели мы используем её для предсказания значений на тестовой выборке (X_test) с помощью метода predict.

4. **Оценка производительности**:
   - Наконец, мы вычисляем среднеквадратичную ошибку (MSE) между предсказанными значениями (y_pred) и фактическими значениями тестовой выборки (y_test). MSE является важным показателем, который помогает оценить качество модели: чем меньше значение MSE, тем лучше модель.


```python
from sklearn import linear_model
import numpy as np
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
coef_list = []
reg_ridge = linear_model.Ridge(alpha=0.1)
reg_ridge.fit(X_train, y_train)
y_pred = reg_ridge.predict(X_test)
print(mean_squared_error(y_test, y_pred), ": ", "0.1")
```

### Обучение модели Ridge регрессии с различными значениями alpha

В этом разделе мы будем обучать модель Ridge регрессии, используя разные значения параметра регуляризации alpha. Ridge регрессия помогает избежать переобучения модели, добавляя штраф за большие значения коэффициентов.

1. **Перебор значений alpha**:
   - Мы будем варьировать значение alpha от 1 до 10. Чем больше значение alpha, тем сильнее регуляризация, что может помочь улучшить обобщающую способность модели, особенно если данные подвержены шуму.

2. **Обучение модели**:
   - Для каждого значения alpha мы создаем экземпляр модели Ridge и обучаем её на обучающей выборке (X_train и y_train).

3. **Прогнозирование и оценка**:
   - После обучения модели мы используем её для предсказания значений на тестовой выборке (X_test).
   - Мы вычисляем среднеквадратичную ошибку (MSE) между предсказанными значениями и фактическими значениями тестовой выборки (y_test). Это позволяет нам оценить качество модели для каждого значения alpha.

4. **Вывод результатов**:
   - Мы выводим среднеквадратичную ошибку для каждого значения alpha, чтобы определить, какое значение обеспечивает наилучшие результаты.


```python
for alpha in np.arange(0.5, 10, 0.5):
  reg_ridge = linear_model.Ridge(alpha=alpha)
  reg_ridge.fit(X_train, y_train)
  coef_list[-1].append()
  y_pred = reg_ridge.predict(X_test)
  print(mean_squared_error(y_test, y_pred), ": ", alpha)
```

Теперь порпробуем более мощную модель, а именно случайный лес.

### Подготовка данных для обучения и тестирования модели

В данном разделе мы будем разделять наш датасет на обучающую и тестовую выборки. Мы используем колонку Year_2019 для определения, какие записи будут относиться к тестовой выборке, а какие — к обучающей.

1. **Создание тестовой выборки**:

   - Затем мы удаляем колонку Mean Score из X_test, чтобы оставить только признаки.

2. **Создание обучающей выборки**:
   - Аналогично, мы выбираем записи, где Year_2019 равно 0, чтобы использовать данные за другие годы для обучения модели.
   - Целевая переменная y_train будет содержать значения Mean Score для обучающей выборки.
   - Мы также удаляем колонку Mean Score из X_train.

3. **Удаление ненужных признаков**:
   - Мы исключаем колонки, относящиеся к годам с 2015 по 2019, из обеих выборок, чтобы сосредоточиться на других признаках.
